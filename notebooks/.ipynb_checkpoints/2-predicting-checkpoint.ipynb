# Bank marketing campaigns dataset | Opening Deposit

It is a dataset that describing Portugal bank marketing campaigns results.
Conducted campaigns were based mostly on direct phone calls, offering bank client to place a term deposit.
If after all marking afforts client had agreed to place deposit - target variable marked 'yes', otherwise 'no'

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt
from sklearn import tree
import plotly.express as px
import plotly.graph_objects as go
from rfpimp import *
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.model_selection import KFold, StratifiedKFold
from tqdm import tqdm
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("../data/clean/bank-additional-full.csv")
```

```python
df.head()
```

```python
X = df.drop("y_yes", axis=1)
#np.random.seed(seed = 42)
# X["random"] = np.random.random(size = len(X))
y = df["y_yes"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
rf_cls = RandomForestClassifier()
rf_cls.fit(X_train, y_train)
y_pred = rf_cls.predict_proba(X_test)
```

```python
score = rf_cls.score(X_test, y_test)
score
```

```python
cross_validate(rf_cls, X_train, y_train, cv=3, scoring=('accuracy', 'average_precision', 'recall','f1'))
```

RandomForestClassifier with out-of-the-box settings has an accuracy of ~0.89.

The problem lies in the low recall score. Low recall score (~0.29) means we found only ~29% of the correct outputs, which means our model is biased to saying "no".

Problem lies in how the accuracy is calculated and the distribution of the output "y_yes" feature. Most of the rows have y_yes value of 0.
If we predict for every row a "rejection", we would have a ~0.87 accuracy.

Rejections=36537, accepted=4639 => 
accuracy = (36537-4639)/36537 => 
accuracy = 0.87303281604

Using ROC Curve (Receiver Operating Characteristic Curve), we can calculate a more realistic accuracy of our model, since it is "class-distribution"-invariant. It's simply a plot of TPR (true positive rate) and FPR (false positive rate)

```python
roc_plot = metrics.plot_roc_curve(rf_cls, X_test, y_test)
```

```python
roc_score = metrics.roc_auc_score(y_test, y_pred[:,1])
print("ROC-AUC score:", roc_score)
```

Let's try to increase the accuracy by optimizing differnet parameters.

```python
skf = StratifiedKFold(shuffle=True, random_state=42)
estimators = [50, 100, 150, 200, 250]

scores = []
for est in tqdm(estimators):
    local_scores = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        classifier = RandomForestClassifier(n_estimators=est)
        classifier.fit(X_train, y_train)
        y_pred = classifier.predict_proba(X_test)
        roc_score = metrics.roc_auc_score(y_test, y_pred[:,1])
        local_scores.append(roc_score)
    scores.append(np.mean(local_scores))    
```

```python
px.line(x=estimators, y=scores)
```

Changing estimators brings us really little improvement. Probably not even worth the additional waiting.

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
max_depths = [5, 10, 15, 20, 25, 30]

scores = []
for param in tqdm(max_depths):
    local_scores = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        classifier = RandomForestClassifier(max_depth=param)
        classifier.fit(X_train, y_train)
        y_pred = classifier.predict_proba(X_test)
        roc_score = metrics.roc_auc_score(y_test, y_pred[:,1])
        local_scores.append(roc_score)
    scores.append(np.mean(local_scores))
```

```python
px.line(x=max_depths, y=scores)
```

Apparently, we can gain some accuracy by choosing the max_depth 10.

This is as good as the researches also got in this research paper: https://github.com/sukanta-27/Predicting-Success-of-Bank-Telemarketing/blob/master/Relevent%20Paper/targeted_marketing.pdf

They used a neural network to achieve 0.8, but they also tested LR, DT, SVM, getting scores between 0.71 and 0.767.

#### Feature importances

```python
def imp_df(column_names, importances):
    df = pd.DataFrame({'feature': column_names,
                       'feature_importance': importances}) \
           .sort_values('feature_importance', ascending = True) \
           .reset_index(drop = True)
    return df
```

```python
rf_cls = RandomForestClassifier(max_depth=10, random_state=42)
rf_cls.fit(X_train, y_train)

base_imp = imp_df(X_train.columns, rf_cls.feature_importances_)
px.bar(base_imp, x="feature_importance", y="feature", orientation="h", height=800)
```

Top 4 most important features using standard importances are: euribor3m, nr_employed, pdays, poutcome_success.

```python
def r2(param0, X_train, y_train):
    return r2_score(y_train, rf_cls.predict(X_train))

imp = permutation_importances(rf_cls, X_test, y_test, r2)
viz = plot_importances(imp, width=12)
viz.view()
```

Top 4 most important features using rfpimp importances are: euribor3m, nr_employed, emp_var_rate, pdays.

Apparently the last contact day of the week matters, which sounds a bit weird. Plotting a histogram of the day_of_week features, gives me a uniform distribution across all days, so I'm not sure why it matters.

Other features euribor3m, age, campaign and other make sense.
